```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plot (3,5,10), Plot the projections on the xy, xz, and yz planes

#target points
px=3
py=5
pz=10

#blank graph
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d') #3d cords
ax.scatter(px,py,pz,color='pink',marker='o',s=100,label='Target Point') #plot og plot

#plot projections on plane
ax.scatter(px,py,0,color='orange',marker='o',s=100,label='Projection on XY Plane') #set z=0
ax.scatter(px,0,pz,color='cyan',marker='o',s=100,label='Projection on XZ Plane') #set y=0
ax.scatter(0,py,pz,color='purple',marker='o',s=100,label='Projection on YZ Plane') #set x=0

#connect lines 
ax.plot3D([px,px],[py,py],[0,pz],color='orange',linestyle='dashed') #line to xy plane
ax.plot3D([px,px],[0,py],[pz,pz],color='cyan',linestyle='dashed') #line to xz plane
ax.plot3D([0,px],[py,py],[pz,pz],color='purple',linestyle='dashed') #line to yz plane 

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Point and its Projections on Coordinate Planes')
ax.legend()
plt.show()
```

```python
#points prof. lee gave me to us
px=3
py=5
pz=10

#3d graph window 
fig= plt.figure() 
ax=fig.add_subplot(111, projection='3d')
ax.scatter(px,py,pz,color='pink', s=100, label='OG Point')
ax.set_title('1st 3D Point and Projections')
#lpane projections, xy plane set z=0
ax.scatter(px, py, 0, color='teal', s=50, label='XY Projection')

#xz plane set y=0
ax.scatter(px, 0, pz, color='purple', s=50, label='XZ Projection')

#yz plane set x=0
ax.scatter(0, py, pz, color='magenta', s=50, label='YZ Projection')

#ax.plot([start_x, end_x], [start_y, end_y], [start_z, end_z])
#only z changes from pz to 0
ax.plot([px,px],[py,py], [pz,0], color='teal', linestyle='dashed')
#only y changes from py to 0
ax.plot([px,px],[py,0], [pz,pz], color='purple', linestyle='dashed')
#only x changes from px to 0
ax.plot([px,0],[py,py], [pz,pz], color='magenta', linestyle='dashed')

# Generate random numbers for the normal vector coordinates
random_array = np.random.randn(3)
A = random_array[0]
B = random_array[1]
C = random_array[2]

# Calculate length manually to normalize the vector to a unit vector
vec_length = np.sqrt(A**2 + B**2 + C**2)
A = A / vec_length
B = B / vec_length
C = C / vec_length

# Set static limits for the axes so the view is clean
ax.set_xlim([-15, 15])
ax.set_ylim([-15, 15])
ax.set_zlim([-15, 15])

# Make a simple grid to draw the plane surface
x_grid = np.linspace(-15, 15, 10)
y_grid = np.linspace(-15, 15, 10)
X, Y = np.meshgrid(x_grid, y_grid)

# Solve for Z using the plane equation: Ax + By + Cz = 0
Z = (-A * X - B * Y) / C
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.3)

#project point onto plane
# Find distance using the dot product formula manually
distance = (px * A) + (py * B) + (pz * C)

# Calculate the coordinates for the new point on the plane
proj_x = px - (distance * A)
proj_y = py - (distance * B)
proj_z = pz - (distance * C)

# Plot the projection point on the plane and draw the connecting line
# Plot the new projection point as a magenta star marker ('*')
ax.scatter(proj_x, proj_y, proj_z, color='orange', marker='*', s=120, label='Plane Projection')
ax.plot([px, proj_x], [py, proj_y], [pz, proj_z], color='orange', linewidth=2)


# --- STEP 5: LABELS AND SHOW ---
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Vector Projections Graph')
ax.legend()

plt.show()
```
